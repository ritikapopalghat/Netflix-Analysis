dropping the unwanted columns:

```python
cols = ['show_id', 'description']
df.drop(cols, axis=1, inplace=True)
df.columns
```

```python
df.head()
```

splitting genres into lists so that our dataframe has only one genre per row for each movie

```python
df['listed_in'] = df['listed_in'].str.split(', ')
df = df.explode('listed_in').reset_index(drop=True)
df.head()
```

```python
df['listed_in'] = df['listed_in'].astype('category')
df['listed_in'].dtypes
```

```python
df.info()
```

TAKING INSIGHTS:

```python
sns.set_style('whitegrid')
```

1. How many types of content are available on Netflix?

```python
df['type'].value_counts()
```

Conclusion: 
Movies dominate the Netflix catalogue in your dataset — they account for ~97% of all content.
TV Shows are very small fraction — only about 3%.
This suggests that in the dataset you’re using, Netflix’s focus appears to be heavily skewed toward movies rather than episodic TV content.
In a real-world context, that’s unusual compared to Netflix today (which actually has a more balanced library), so
a. Either the dataset may not be up to date.
b. It might represent a specific period or selection.

2. Which year had the highest number of content release?

```python
df['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(20,4), title='Releases per year:')
```

Conclusion:
Content releases grew steadily from the early 2010s, peaked around 2017–2019, and show a slight decline afterward, indicating Netflix's aggressive expansion period followed by stabilization.

3. What is the distribution of content across countries (top 10)? 

```python
df['country'].value_counts().head(10).plot(kind='bar', title='Top 10 Countries by Content Count', color='green')
```

Conclusion:
The United States produces the largest share of Netflix content, followed by India and other countries showcasing the dominance of Hollywood and showing the apaptation of regional content.

4. which genres are most common on Netflix?

```python
df['listed_in'].str.split(', ').explode().value_counts().head(10).plot(kind='pie', title='Top 10 Genres', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

5.Has Netflix increased international content recently?

```python
country_trend = df.groupby('release_year')['country'].nunique()
country_trend.plot(kind='bar', figsize=(20,4), title='Unique Countries per Year', color='black')
```

6.Countries that specialize in particular genre

```python
genre_split = df.assign(genre=df['listed_in'].str.split(', ')).explode('genre')
country_genre_counts = genre_split.groupby(['country', 'genre']).size().reset_index(name='count')
country_genre_counts.sort_values('count', ascending=False).head(30)
```

7.Countries producing mainly mature audience content.

```python
country_rating_counts = df.groupby(['country', 'rating']).size().reset_index(name='count')
tv_ma_by_country = country_rating_counts[country_rating_counts['rating'] == 'TV-MA'] \
    .sort_values('count', ascending=False)
tv_ma_by_country.head(40)
```
